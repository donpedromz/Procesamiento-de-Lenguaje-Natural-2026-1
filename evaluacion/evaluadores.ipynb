```python
def evaluar_clasificador(modelo, test_docs, test_labels):

    predicciones = [modelo.clasificar(doc)[0] for doc in test_docs]

    clases = list(set(test_labels))

    metricas = {}

    for clase in clases:
        VP = sum(p == clase and r == clase for p, r in zip(predicciones, test_labels))

        FP = sum(p == clase and r != clase for p, r in zip(predicciones, test_labels))

        FN = sum(p != clase and r == clase for p, r in zip(predicciones, test_labels))

        prec = VP / (VP + FP) if VP + FP > 0 else 0

        recall = VP / (VP + FN) if VP + FN > 0 else 0

        f1 = 2 * prec * recall / (prec + recall) if prec + recall > 0 else 0

        metricas[clase] = {"precision": prec, "recall": recall, "f1": f1}

    acc = sum(p == r for p, r in zip(predicciones, test_labels)) / len(test_labels)

    return acc, metricas
```
